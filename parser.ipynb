```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://all-events.ru/events/calendar/city-is-moskva/type-is-conferencia/"

response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    # Поиск контейнера, который содержит информацию о конференциях
    conferences = soup.find_all('div', class_='event_flex_item')

    # Перебор найденных конференций
    for conference in conferences:
        # Извлечение названия
        title = conference.find('a', class_='event_name_new')
        title = title.text.strip() if title else 'Нет названия'
        
        # # Извлечение даты
        # date = conference.find('span', class_='conference-date') 
        # date = date.text.strip() if date else 'Нет даты'
        
        # # Извлечение тегов
        # tags = conference.find_all('span', class_='conference-tag')
        # tags = [tag.text.strip() for tag in tags] if tags else ['Нет тегов']


        print(f"Название: {title}")
        # print(f"Дата: {date}")
        # print(f"Теги: {', '.join(tags)}")
        print('-' * 30)
else:
    print(f"Ошибка при получении страницы: {response.status_code}")
```
